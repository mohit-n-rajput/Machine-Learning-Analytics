## Set another column of the dataframe as the index. We will use the set_index function. 

You can read more about here: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.set_index.html'***

---

````python
data.head()
````

 ##### <span class="mark">change the index of the dataframe</span>
 
***drop=True is used to drop the column that's set as index.It is very helpful in the case of duplicate columns.***

***inplace=True is used to make changes in the original dataframe.***


#### What if drop = False

````python
data = pd.read_csv("datasets/big_mart_sales.csv")
````

````python
data.set_index('Outlet_Establishment_Year', drop=False, inplace=True)
data.head()
````

###### <span class="mark">As we see above establishment year is not drop from columns , so there area two same columns. In this case, drop=True comes to help.</span>

````python
data.set_index('Outlet_Establishment_Year', drop=True,inplace = True)
data.head()
````

###### <span class="burk">If inplace = False then index is not updated to dataframe. if you want to change dataframe inplace = True is must. If inplace= False then dataframe index are again default to 0,1,2,3,..</span> 

````python
data = pd.read_csv("datasets/big_mart_sales.csv")
````

````python
data.set_index('Outlet_Establishment_Year', drop=True, inplace=False)
````

###### Since ```inplace = False``` after some time changes are gone and dataframe is  not updated and index are set to default.

````python
data.head()
````

````python
data.index
````

#### Use of inplace parameter.

````python
data = pd.read_csv("datasets/big_mart_sales.csv")

data.set_index('Outlet_Establishment_Year', drop=True, inplace=True)
````

<span class="mark">Now, establishment year column dropped and index also updated.</span>

````python
data.head()
````

##### So drop and inplace parameters are very helpful.

<span class="mark">

<b>When you use ```inplace = True``` parameter, then pandas create a new datafrane with changes you made and  returns a new Dataframe to you.</b>

<b>Therefore, It is advised, that when you use ```inplace = True```, Save a changed dataframe to a new variable. So, original data are not changed and remain safe.</b>
    

</span>

````python
data = pd.read_csv("datasets/big_mart_sales.csv")
````

````python
# 1. change the index of the dataframe
# 2. drop=True is used to drop the column that's set as index
# 3. inplace=True is used to make changes in the original dataframe
# data = pd.read_csv('datasets/big_mart_sales.csv')

new_data = data.set_index('Outlet_Establishment_Year', drop=True, inplace=True)
````

````python
new_data.head()
````

<b>If we store it in  a new variable then set the ```inplace = False```. </b>


````python
data = pd.read_csv("datasets/big_mart_sales.csv")
````

````python
# 1. change the index of the dataframe
# 2. drop=True is used to drop the column that's set as index
# 3. inplace=True is used to make changes in the original dataframe
# data = pd.read_csv('datasets/big_mart_sales.csv')

new_data = data.set_index('Outlet_Establishment_Year', drop=True, inplace=False)
````

````python
new_data.head()
````

````python
# example of a label based index
data.head()
````

````python
new_data.index
````
