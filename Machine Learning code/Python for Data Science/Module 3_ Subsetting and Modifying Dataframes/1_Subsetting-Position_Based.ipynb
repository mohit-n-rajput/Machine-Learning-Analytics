#### `READ THE DATA`

- In this notebook, we are going to use the big mart sales data that we have used previously. It is stored in the folder name `datasets`.

```python
# importing the pandas library
import pandas as pd
```

```python
data = pd.read_csv('datasets/big_mart_sales.csv')
```

## `How to view the top and bottom rows of the data?`

- Use `head` function to view top n rows.
- Use `tail` function to view bottom n rows.

---

```python
# view top 2 rows
data.head(2)
```

```python
# view top 10 rows
data.head(10)
```

```python
# view top  5 rows by default
data.head()
```

```python
# view bottom 2 rows
data.tail(2)
```

```python
data.tail()
```

## `How to select rows in a particular range?`

---

```python
# select the data from range 10-15
data[10:15]
```

```python
data[0:10]
```

```python
data[0:21:2]
```

```python
# negative index:
data[-10::-1]
```

## `How to select the rows by position?`

- When we are using `iloc`, we need to specify the rows and columns by their position.
---

DOC: data.iloc([row numbers], [column numbers])

***iloc is not function it is a attribute of dataframe itself.***

```python
data.iloc([i for i in range(0,10,2)])
```

```python
# select specific rows by index number
data.iloc[[1,5,2,4,6,14]]
```

#### `How to select the specific rows and columns from the data using their position?`

- <span class="mark">In the `iloc` function pass the first list as the order of rows by their index and pass the second list as the order of columns.</span>

---

```python
data.iloc[[1,4,5,2],[1,3,5,89]]
```

```python
data.iloc[[1,4,5,2],[1,3,5]]
```

```python
data.iloc[[1,4,5,2],[1,3,5,8]]
```

```python
# same as slicing operation above
data.iloc[1:16]
```

```python
data.iloc[::-1]
```

```python
data.iloc[10:20:2]
```
